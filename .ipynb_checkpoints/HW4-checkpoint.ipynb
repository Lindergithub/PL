```python
import bs4
soup=bs4.BeautifulSoup(data, "html.parser")#分析html，所以要安裝第三方套件：BeautifulSoup4
politics=soup.find("div", class_="container-fluid markdown-body")#政見politics＝div標籤其中class等於"container-fluid markdown-body"
```

```python
str=politics.string#只印政見
print(str)
```

```python
#用re把每人的政見找出來
import re
content=[]
str1=re.sub("(\\n)", " ", str)#以空白取代"\n"
strl2=re.sub("(#{3})", "%%", str1)
#文章為markdown語法，想要留下###第幾輪、####誰說的，但之前的部分不知道要怎麼刪，所以先想辦法隔開他們
strl3=re.sub(" ", "",strl2)
strl3=re.sub(" ", "",strl3)#刪掉空白
regex=re.compile(r'(\s)|(\\u+[0-9]{4})')#正則化表達\u3000
strl4=re.sub(regex, "",strl3)#把\u3000刪掉
strl5=re.sub("#", "", strl4)
#之前把###變%、把####變%#，現在把所有#都刪掉，因為#跟##的內容不須要，%#變%也比較簡單之後不用繼續刪#
strl6=re.split("%%", strl5)
strl6
```

```python
import pandas as pd
#做成好看的dataframe
times = []
name = []#候選人
content = []#逐字稿
la = []#剩下的
de = []#不要的

for con in strl6:
    data = con
    data=re.sub(" ", "", data)
    regex_name = re.compile(r'(蔡英文)：|(韓國瑜)：|(宋楚瑜)：')
    match_name = regex_name.search(data)
    if match_name != None:
        name.append(match_name.group())#name完成
        data1 = data.replace(match_name.group(), "") 
        content.append(data1)#content完成
    else:
        la.append(data)
        
for pon in la:
    dat = pon
    regex = re.compile(r'第')
    match = regex.search(dat)
    if match != None:
        times.append(dat)
    else:
        de.append(dat)
#因為times輸出會是：['第一輪---', '第二輪', '第三輪---']，真的醜，但很簡單所以我直接手改
times=("第一輪","第一輪","第一輪","第二輪","第二輪","第二輪","第三輪","第三輪","第三輪")#times完成
print(name)
df = pd.DataFrame(
    {'輪': times,
     '候選人': name,
     '逐字稿': content
    })
df
```

```python
#發現同一候選人說的要放一起後面才好處理，所以df要重做ohmygod好麻煩
name1=[]
content1=[]
name2=[]
content2=[]
name3=[]
content3=[]
for x in strl6:
    data = x
    data=re.sub(" ", "", data)
    regex_name1 = re.compile(r'(蔡英文)：')#|(韓國瑜)：|(宋楚瑜)：
    match_name1 = regex_name1.search(data)
    regex_name2 = re.compile(r'(韓國瑜)：')#|(韓國瑜)：|(宋楚瑜)：
    match_name2 = regex_name2.search(data)
    regex_name3 = re.compile(r'(宋楚瑜)：')#|(韓國瑜)：|(宋楚瑜)：
    match_name3 = regex_name3.search(data)
    if match_name1 != None:
        name1.append(match_name1.group())
        data1 = data.replace(match_name1.group(), "") 
        content1.append(data1)
    if match_name2 != None:
        name2.append(match_name2.group())
        data2 = data.replace(match_name2.group(), "") 
        content1.append(data2)
    if match_name3 != None:
        name3.append(match_name3.group())
        data3 = data.replace(match_name3.group(), "") 
        content1.append(data3)

name=(name1+name2+name3)
content=(content1+content2+content3)
df1 = pd.DataFrame(
    {'候選人': name,
     '逐字稿': content
    })
df1

#content1=[1,2,3]
#[1,2,3,4,5,6,7,8,9]
#[123,456,789]
#[1,2,3]
```
